OUTLIERS IN SALES AND PROFIT

```python
Q1 = df[['Sales','Profit']].quantile(0.25)
Q3 = df[['Sales','Profit']].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers = df[((df['Sales'] < lower['Sales']) | (df['Sales'] > upper['Sales']))
              | ((df['Profit'] < lower['Profit']) | (df['Profit'] > upper['Profit']))]

print("Actual outliers count:", outliers.shape[0])
print("Outliers in Sales:", outliers['Sales'].shape[0])
print("Outliers in Profit:", outliers['Profit'].shape[0])   
```

Sales: Some very high sales values (e.g., large orders) appear as outliers.

Profit: Negative profits (losses) and high profit values occur — these often register as outliers in EDA

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

DISCOUNT VS PROFIT

```python
sns.scatterplot(x='Discount', y='Profit', data=df)
print(df['Discount'].corr(df['Profit']))
plt.show()
```

Usually a negative correlation — higher discounts often correspond with lower profit.

SALES VS QUANTITY

```python
sns.scatterplot(x='Quantity', y='Sales', data=df)
print(df['Sales'].corr(df['Quantity']))
```

Positive correlation — more units sold generally leads to higher total sales.
