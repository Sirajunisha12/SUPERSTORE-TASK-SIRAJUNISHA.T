Load the dataset

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel(r"c:\Users\ADMIN\OneDrive\Desktop\sirajunisha super store dataset.xlsx")
```

 Remove duplicate rows 

```python
duplicates_count = df.duplicated().sum()
print(f"Duplicates found: {duplicates_count}")
df = df.drop_duplicates()
print("Duplicates removed.")
```

Standardize date formats

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], errors='coerce')

# Output examples
print("Order Date types:", df['Order Date'].dtype)
print(df[['Order Date', 'Ship Date']].head())
```

Handle missing value

```python
missing_summary = df.isnull().sum()
print("Missing values before handling:")
print(missing_summary)
```

 Fix inconsistent category / segment values 

```python
df['Segment'] = df['Segment'].str.strip().str.title()
df['Category'] = df['Category'].str.strip().str.title()
df['Sub-Category'] = df['Sub-Category'].str.strip().str.title()

# Example checks
print("Unique segments:", df['Segment'].unique())
print("Unique categories:", df['Category'].unique())
```

Convert numerical fields properly

```python
# Already converted above, ensure types
df['Sales'] = df['Sales'].astype(float)
df['Profit'] = df['Profit'].astype(float)
df['Quantity'] = df['Quantity'].astype(int)

print("Data types after numeric conversion:")
print(df[['Sales','Profit','Quantity']].dtypes)
```

Save cleaned datase

```python
df.to_csv("cleaned_superstore.csv", index=False)
print("Cleaned dataset saved as cleaned_superstore.csv")
```
